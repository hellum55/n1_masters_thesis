```python
import os
import shutil
import pandas as pd
```

```python
# Paths
csv_file = "C:/Users/Christian/Desktop/sampled_dataset.csv"
source_folder = "C:/Users/Christian/Desktop/downloadet_pictures_2"
destination_folder = "C:/Users/Christian/Desktop/image_data"

# Create destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Read image filenames from the "ImageName" column
df = pd.read_csv(csv_file)
image_names = df["ImageName"].tolist()

# Copy only the desired images
copied = 0
for image_name in image_names:
    source_path = os.path.join(source_folder, image_name)
    if os.path.exists(source_path):
        shutil.copy(source_path, destination_folder)
        copied += 1
    else:
        print(f"❌ Image not found: {image_name}")

print(f"✅ Done. {copied} images copied to {destination_folder}")
```

```python
# Filter rows where the image file exists in the source folder
df_filtered = df[df["ImageName"].apply(lambda img: os.path.exists(os.path.join(source_folder, img)))]

# Save the filtered DataFrame back to the CSV file
df_filtered.to_csv(csv_file, index=False)

print(f"✅ Updated CSV file saved with {len(df_filtered)} rows.")
```

```python
import pandas as pd
import os
import shutil

# Set your paths
excel_file = 'C:/Users/Christian/Desktop/N1_data/deleted_images.xlsx'  # replace with your actual file name
image_folder = 'C:/Users/Christian/Desktop/N1_data/image_data'
deleted_folder = 'C:/Users/Christian/Desktop/N1_data/deleted_images'

# Create deleted_images folder if it doesn't exist
os.makedirs(deleted_folder, exist_ok=True)

# Load the image IDs (assuming there's a column called 'image_id')
df = pd.read_excel(excel_file)
image_ids = df['id'].tolist()

# Move each image
for image_id in image_ids:
    image_name = f"{image_id}.png"
    source_path = os.path.join(image_folder, image_name)
    destination_path = os.path.join(deleted_folder, image_name)

    if os.path.exists(source_path):
        shutil.move(source_path, destination_path)
        print(f"Moved: {image_name}")
    else:
        print(f"File not found: {image_name}")
```

```python
import pandas as pd

# Load df_filtered CSV
df_filtered = pd.read_csv('C:/Users/Christian/Desktop/N1_data/df_filtered.csv')

# Load deleted image IDs from Excel file
df_deleted = pd.read_excel('C:/Users/Christian/Desktop/N1_data/deleted_images.xlsx')

# Strip '.png' from ImageName column if needed
df_filtered['ImageName'] = df_filtered['ImageName'].str.replace('.png', '', regex=False)

# Get list of IDs from deleted_images.xlsx
deleted_ids = df_deleted['id'].astype(str).tolist()

# Filter df_filtered to exclude deleted IDs
df_filtered_updated = df_filtered[~df_filtered['ImageName'].isin(deleted_ids)]

# Save the cleaned DataFrame
df_filtered_updated.to_csv('df_filtered_updated.csv', index=False)

print(f"Filtered DataFrame saved with {len(df_filtered_updated)} rows.")
```

```python
pivot_table = df_filtered_updated.pivot_table(index='ValidatedType', columns='ValidatedConditionId', aggfunc='size', fill_value=0)
pivot_table['Total'] = pivot_table.sum(axis=1)
pivot_table
```
